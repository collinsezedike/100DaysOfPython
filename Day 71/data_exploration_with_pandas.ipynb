```python
highest_mid_career_salary_index = clean_df["Mid-Career Median Salary"].idxmax()
highest_paying_major = clean_df.loc[highest_mid_career_salary_index]["Undergraduate Major"]
highest_paying_major_salary = clean_df.loc[highest_mid_career_salary_index]["Mid-Career Median Salary"]

lowest_starting_salary_index = clean_df["Starting Median Salary"].idxmin()
lowest_starting_pay_major = clean_df.loc[lowest_starting_salary_index]["Undergraduate Major"]
lowest_starting_pay_major_salary = clean_df.loc[lowest_starting_salary_index]["Starting Median Salary"]

lowest_mid_career_salary_index = clean_df["Mid-Career Median Salary"].idxmin()
lowest_paying_major = clean_df.loc[lowest_mid_career_salary_index]["Undergraduate Major"]
lowest_paying_major_salary = clean_df.loc[lowest_mid_career_salary_index]["Mid-Career Median Salary"]

print(f"Highest paying mid career major is {highest_paying_major} with a salary of {highest_paying_major_salary}")
print(f"Lowest starting pay major is {lowest_starting_pay_major} with a salary of {lowest_starting_pay_major_salary}")
print(f"Lowest paying mid career major is {lowest_paying_major} with a salary of {lowest_paying_major_salary}")
```

```python
spread = clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
clean_df.insert(1, "Spread", spread)
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Spread")
low_risk[["Undergraduate Major", "Spread"]].head()
```

```python
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
highest_potential.head()
```

```python
highest_spread = clean_df.sort_values("Spread", ascending=False)
highest_spread.head()
```
